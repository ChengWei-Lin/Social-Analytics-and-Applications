```
# Imports the Google Cloud client library
from google.cloud import language
from google.cloud.language import enums
from google.cloud.language import types
import os
from google.oauth2 import service_account

credentials = service_account.Credentials.from_service_account_file('/content/drive/My Drive/IS434-5294bc3e3a83.json')

from google.cloud import language
# Instantiates a client
client = language.LanguageServiceClient(credentials=credentials)
```

```
sentiment_list = []
```

```
for text in df2['comments']:
  document = types.Document(
      content=str(text),
      type=enums.Document.Type.PLAIN_TEXT)

  # Detects the sentiment of the text
  try:
    sentiment = client.analyze_sentiment(document=document).document_sentiment
    sentiment_list.append((sentiment.score, sentiment.magnitude))
  except:
    sentiment_list.append((0,0))

```

```
print(sentiment_list)
```

```
len(sentiment_list)
```

```
score = []
magnitude = []

for i in range(len(sentiment_list)):
  score.append(sentiment_list[i][0])
  magnitude.append(sentiment_list[i][1])
```

```
print(len(df2))
print(len(score))
```

```
df2['google_score'] = score
```

```
from textblob import TextBlob

textblob_score = []

for sentence in df2['comments']:
  sentiment = TextBlob(str(sentence)).sentiment
  textblob_score.append(sentiment.polarity)
```
